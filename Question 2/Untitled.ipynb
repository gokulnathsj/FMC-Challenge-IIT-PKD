```python
import numpy as np 
import pandas as pd
from keras.layers import Dense
from keras.utils import to_categorical
from keras.models import Sequential
import pickle
```

```python
data = pd.read_csv("fashion-mnist_train.csv")
data_test = pd.read_csv("Q2_Clothing_test.csv")
```

```python
data
```

```python
X_train = data.iloc[:, 1:].values
y_train = data.iloc[:, 0].values
X_test = data_test.iloc[:,:].values
```

```python
X_train = X_train / 255
X_test  = X_test / 255
y_train = y_train 
```

```python
model = Sequential()
model.add(Dense(128, activation='relu', input_dim=784))
model.add(Dense(128, activation='relu'))
model.add(Dense(10,  activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train, to_categorical(y_train, 10), epochs = 100, batch_size = 20)
```

```python
model.save_weights('Question2.h5')
```

```python
y_pred = model.predict(X_test)
y_pred = np.argmax(y_pred, axis = 1)
```

```python
f = open('Test_set_results.txt','w')
for i in y_pred:
    if i == 0:
        f.write("T-shirt/top \n")
    elif i == 1:
        f.write("Trouser \n")
    elif i == 2:
        f.write("Pullover \n")
    elif i == 3:
        f.write("Dress \n")
    elif i == 4:
        f.write("Coat \n")
    elif i == 5:
        f.write("Sandal \n")
    elif i == 6:
        f.write("Shirt \n")
    elif i == 7:
        f.write("Sneaker \n")
    elif i == 8:
        f.write("Bag \n")
    elif i == 9:
        f.write("Ankle boot \n")
f.close()
```
